```python
import requests
from bs4 import BeautifulSoup
```

# Requests

```python
url = 'https://www.ijunoon.com/dictionary/cut-urdu-meaning/'
```

```python
response = requests.get(url)
print(response)
print(response.content)
```

# BeautifulSoup

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
urdu_results = soup.find_all('div', class_='urdu_results', id='urdu_results')
print(urdu_results[0])
print(len(urdu_results))
```

```python
urdu = urdu_results[0].find_all('div', class_='u')
roman = urdu_results[0].find_all('div', class_='r')
print(urdu)
print(roman)
```

```python
urdu_words = []
for e in urdu:
    urdu_words.append(e.text)
```

```python
urdu_words = [ e.text for e in urdu ]
print(urdu_words)
```

```python
roman_words = [ e.text for e in roman ]
print(roman_words)
```

```python
data = list(zip(urdu_words, roman_words))
print(type(data))
for d in data:
    print(d)
```

```python
with open('data.txt', 'w', encoding='utf-8') as file:
    for d in data:
        file.write(str(d))
```
