```python
import json
import requests
from bs4 import BeautifulSoup
```

```python
old_url = 'https://rekhta.org/poets'
head_url = 'https://rekhta.org/PoetCollection?'
lang = 'lang='
page = '&pageNumber='
starts = '&Info=poet&StartsWith='
tail_url = '&keyword=&typeID=659186cb-44e7-4d94-8b1a-fc70f939a733&TypeSlug=poets&_=1589285769078'


'https://rekhta.org/PoetCollection?lang=1&pageNumber=10&Info=poet&StartsWith=a&keyword=&typeID=659186cb-44e7-4d94-8b1a-fc70f939a733&TypeSlug=poets&_=1589285769086'
```

```python
poets = []
for c in 'abcde':
    for p in range(51):
        url = head_url + lang + '3' + page + str(p) + starts + c + tail_url
        response = requests.get(url)
        soup = ''
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            listing = soup.find_all('div', class_='poetNameDatePlace clearfix')  
            if len(listing) > 0:
                for item in listing:
                    link = item.find_all('a')[0]['href']
                    name = item.find_all('h2')[0].text
                    poets.append([link,name])
            else:
                break
print(poets)
```

```python
with open('poet-listing.json', 'w', encoding='utf-8') as file:
    json.dump(poets, file, ensure_ascii=False, indent=2)
```
